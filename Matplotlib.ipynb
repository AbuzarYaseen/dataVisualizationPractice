```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

**Matplotlib Basics**

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)   # 1st argument will be for x-axis and 2nd will be for y-axis
plt.title("View Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

**Subplots in Matplotlib**

**Means creating multiple plots together.**

```python
plt.subplot(2,3,1)
plt.plot(y**2,x)
plt.subplot(2,3,4)
plt.plot(x,y)
plt.subplot(2,3,3)
plt.plot(y,x)
plt.subplot(2,3,2)
plt.plot(x**2,x**2)
```

**Object Oriented Way**

Values describe how much you wants to add space in the plot from left and bottom respectivly. And last 2 values defines how much will be the width(0.7 means 70%) and height(0.8 means 80%)

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(7,4),dpi=100)
axis1 = fig.add_axes([0.1, 0.1, 0.7, 0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.6, 0.2, 0.4, 0.3])
axis2.plot(x,y)

```
