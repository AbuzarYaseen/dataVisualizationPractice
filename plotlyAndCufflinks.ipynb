```python
import seaborn as sns
import pandas as pd 
import cufflinks as cf 
from plotly.offline import iplot
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
fig = px.line(tips, y='total_bill')
fig.show()
```

```python
avg_tip = tips.groupby('day', observed=False)['tip'].mean().reset_index()

fig = px.bar(avg_tip, x='day', y='tip')
fig.show()
```

```python
fig = px.scatter(tips, x='total_bill', y='tip', color='day')
fig.show()
```

```python
avg_tip = tips.groupby('total_bill', observed=False)['day'].mean().reset_index()

fig = px.bar(avg_tip, x='total_bill', y='day')
fig.show()
```
